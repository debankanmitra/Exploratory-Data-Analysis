# <center> Feature Engineering </center>

WE WILL BE PERFORMING THE FOLLOWING STEPS IN FEATURE ENGINEERING
1) MISSING VALUES <br>
2) TEMPORAL VARIABLES <br>
3) CATEGORICAL VARIABLES : REMOVE RARE LABELS <br>
4) STANDARDISE THE VALUES OF THE VARIABLES TO THE SAME RANGE 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns',None)
dataset=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Advanced-House-Price-Prediction-/master/train.csv')
dataset.head()
```

## Missing Values

```python
# Categorical features with missing values
categorical_nan=[feature for feature in dataset.columns if dataset[feature].dtype=='O' and dataset[feature].isnull().sum()>1]
for i in categorical_nan:
    print(i,'  ',np.round((dataset[i].isnull().sum()/1460)*100,2),'% missing')
    
```

```python
# Replace missing values with new label
data=dataset.copy()
data[categorical_nan]=data[categorical_nan].fillna('Missing')

data[categorical_nan].isnull().sum()    
```

```python
# Numerical features with missing values
numerical_nan=[feature for feature in dataset.columns if dataset[feature].dtype!='O' and dataset[feature].isnull().sum()>1]
for i in numerical_nan:
    print(i,'  ',np.round((dataset[i].isnull().sum()/1460)*100,2),'% missing')
```

```python
# Replace missing values with median / mode
## when we have lot of outliers , we should definitely replace the values with median/mode
for i in numerical_nan:
    dataset[i].fillna(dataset[i].median(),inplace=True)

dataset[numerical_nan].isnull().sum()     
```

```python
dataset[numerical_nan].head()
```

```python
# Temporal variables (Date time variables)
# we are calculating the age for all the features
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()    
```

# Numerical Variables

Since the numerical variables are skewed we will perform log normal distribution

```python
dataset.head()
```

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
dataset.head()    
```

# Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_features
```

```python
dataset.groupby("MSZoning")['SalePrice'].count()
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
dataset.head(10)
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
dataset.head(10)    
```

```python
dataset["MSZoning"].unique()
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)
```

```python
dataset.head()
```

# Feature Scaling


```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
scaler.transform(dataset[feature_scale])
```

```python
dataset.head()
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```
